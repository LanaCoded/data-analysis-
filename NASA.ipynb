```python
# does type of star effect the type of plannet
# Create a Star Type category based on temperature
def categorize_star(temp):
    if temp < 3700: return 'M-Dwarf (Cool)'
    elif temp < 5200: return 'K-Type'
    elif temp < 6000: return 'G-Type (Sun-like)'
    else: return 'F/A-Type (Hot)'

df_clean['star_type'] = df_clean['koi_steff'].apply(categorize_star)

# Visualize Star Type vs Planet Radius
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 6))
sns.boxplot(data=df_clean[df_clean['koi_disposition'] == 'CONFIRMED'],
            x='star_type', y='koi_prad', palette='magma')
plt.yscale('log')
plt.title('Does Star Temperature Affect Planet Size?')
plt.ylabel('Planet Radius (Earth Radii)')
plt.show()
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier

models = {
    'Logistic Regression': LogisticRegression(max_iter=1000),
    'Random Forest': RandomForestClassifier(),
    'Support Vector Machine': SVC(),
    'Decision Tree': DecisionTreeClassifier(),
    'Gradient Boosting': GradientBoostingClassifier(),
    'K-Nearest Neighbors': KNeighborsClassifier()
}
```

```python
from sklearn.metrics import accuracy_score

# Dictionary to store the results
results = {}

print("🚀 Starting Model Training...\n")

for name, model in models.items():
    # 1. Train the model
    model.fit(X_train, y_train)

    # 2. Make predictions
    predictions = model.predict(X_test)

    # 3. Calculate accuracy
    accuracy = accuracy_score(y_test, predictions)

    # 4. Store and print result
    results[name] = accuracy
    print(f"{name}: {accuracy:.4f}")

print("\n✅ All models trained!")
```

```python
# Sort models by accuracy (highest first)
sorted_results = sorted(results.items(), key=lambda x: x[1], reverse=True)

print("🏆 Model Rankings:")
for name, score in sorted_results:
    print(f"{name}: {score*100:.2f}%")
```
